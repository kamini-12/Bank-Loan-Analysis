```python
import pandas as pd
import numpy as np
application = pd.read_csv('application_data.csv')
application
df = pd.DataFrame(application)
import matplotlib.pyplot as plt
```

```python
gender_counts = df['TARGET'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%')
plt.title('LATE PAYMENT AND TIMELY PAYMENTS')
plt.legend()
plt.show()
```

```python
# value_counts = df.groupby('NAME_CONTRACT_TYPE','TARGET')['0'].count()
value_counts = df.groupby(['NAME_CONTRACT_TYPE','TARGET']).size()

group_keys = [str(key) for key in value_counts.index]

# Create a bar chart
plt.bar(group_keys, value_counts.values,width=0.25)
plt.title('Data Dependent on Two Columns')
plt.xlabel('Column1 and Column2')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
# import modules
import matplotlib.pyplot as mp
import pandas as pd
import seaborn as sb
import numpy as np

# import file with data
data = pd.read_csv('application_data.csv')
filtered_data = df[df['TARGET'] == 0]

# creating mask
mask = np.triu(np.ones_like(filtered_data.corr()))

# plotting a triangle correlation heatmap
dataplot = sb.heatmap(data.corr(), mask=mask)
mp.title('Late payers')

# displaying heatmap
mp.show()
```
